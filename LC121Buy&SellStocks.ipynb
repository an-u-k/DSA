## LC 121 Best Time to sell and Buy stocks

Time complexity -O(n)

Space complexity -O(1)

Approach:
a) setting the minimum profit to infinity 
b) loops throgh each price to find the max profit 

```python
from typing import List
```

```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        profit = 0
        min_price = float('inf')
        for price in prices:
            min_price = min(min_price, price)
            if price - min_price > profit:
                profit = price - min_price
        return profit
```

```python
sol = Solution()
print(sol.maxProfit([7,5,4]))
```

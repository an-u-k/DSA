## LC344 Reverse String

Out-space array : 
    we can use s[::-1]
    The problem with this is it creates a new string 

In-place array :
    we have to reverse the string without creating extra memory
    reverse in-place → no new list/string
               → O(1) space
               → modify original directly

```python
from typing import List
```

```python
class Solution:
    def reverseString(self, s: List[str]) -> None:
        left = 0 
        right = len(s) - 1
        while left < right:
            s[left], s[right] = s[right], s[left]
            left += 1
            right -= 1
```

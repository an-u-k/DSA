## LC217- Contains duplicate 

Approach :
Coverts the list to a set 
Set in python contains only unique elements unlike list 

Space complexity -O(n)

Time complexity - O(n)


```python
from typing import List
```

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        array_unique = set(nums)
        if len(array_unique) == len(nums):
            return False
        else:
            return True
    
```

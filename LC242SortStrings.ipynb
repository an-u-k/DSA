```python
## LC217- SortString.

Approach:
Sort both strings and check if they are equal.
This is easy but the time complexity is O(n logn) which is not efficient.

So i used Hashmap(dict) to reduce the time complexity 
Time complexity -O(n).
Space Complexity- O(n)

```

```python
class Solution1:
    def isAnagram(self, s: str, t: str) -> bool:
        s_sorted = sorted(s)
        t_sorted = sorted(t)
        if(s_sorted ==  t_sorted):
            return True
        return False
```

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        count ={}
        for char in s:
            count[char]=count.get(char,0) +1

        for char1 in t:
            count[char1]=count.get(char1,0) -1

        for val in count.values():
            if val!= 0:
                return False
    
        return True

```

```python
sol = Solution()
print(sol.isAnagram("anagram", "nagaram"))  
print(sol.isAnagram("rat", "car"))          
print(sol.isAnagram("hello", "world"))  
```
